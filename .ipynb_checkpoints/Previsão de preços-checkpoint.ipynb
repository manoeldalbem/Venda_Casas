```python
# função que calcula o erro percentual absoluto
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mape
```

```python
kf = KFold(n_splits=10)
outcomes_r2 = []
outcomes_mape = []
out_mean_squared_log_error = []

    
fold = 0
for train_index, test_index in kf.split(X):
    fold += 1
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    model_best.fit(X_train, y_train)
    ypred = model_best.predict(X_test)
    
    mape = mean_absolute_percentage_error(y_test, ypred)*100
    score = r2_score(y_test, ypred)*100
    log_error = np.sqrt(mean_squared_log_error(y_test, ypred))
    
    outcomes_mape.append(mape)
    outcomes_r2.append(score)
    out_mean_squared_log_error.append(log_error)
    
    print(f"Fold {fold} MAPE: {mape}, R2: {score}, log error: {log_error}")     

mean_mape = np.mean(outcomes_mape)
mean_r2 = np.mean(outcomes_r2)
mean_log_error = np.mean(out_mean_squared_log_error)

print("\nMean Mape: {0}".format(mean_mape))
print("\nMean R2: {0}".format(mean_r2))
print("\nMean Roo log error: {0}".format(mean_log_error))
```
